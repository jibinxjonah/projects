# anova

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols 
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy.stats import chi2_contingency
```

```python
model=ols('height ~ cardio',data=df).fit()
anova=sm.stats.anova_lm(model,type =2)
print(anova.round(decimals=5))
```

```python
model=ols('weight ~ cardio',data=df).fit()
anova=sm.stats.anova_lm(model,type =2)
print(anova.round(decimals=5))
```

```python
model=ols('ap_hi ~ cardio',data=df).fit()
anova=sm.stats.anova_lm(model,type =2)
print(anova.round(decimals=5))
```

```python
model=ols('ap_lo ~ cardio',data=df).fit()
anova=sm.stats.anova_lm(model,type =2)
print(anova.round(decimals=5))
```

```python
model=ols('cholesterol ~ cardio',data=df).fit()
anova=sm.stats.anova_lm(model,type =2)
print(anova.round(decimals=5))
```

# splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('cardio',axis=1)
```

```python
y=df.cardio
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
df.head()
```

```python
xtrain.iloc[:,1:5]  = sc.fit_transform(xtrain.iloc[:,1:5])
xtest.iloc[:,1:5]  = sc.transform(xtest.iloc[:,1:5])
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
```

```python
model=lo.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```
